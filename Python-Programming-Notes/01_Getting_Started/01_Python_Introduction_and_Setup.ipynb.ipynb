###### C++ vs Java vs Python

####### Comparison Table

| Feature       | C++                                | Java                               | Python                             |
|---------------|------------------------------------|------------------------------------|------------------------------------|
| Type          | Compiled language                  | Compiled to bytecode, runs on JVM  | Interpreted language               |
| Syntax        | Complex, many rules                | Strict, everything in classes       | Very simple, uses indentation      |
| Speed         | Fastest (directly machine code)    | Medium (runs on JVM, adds a layer) | Slowest (interpreted line by line) |
| Memory        | Manual (new/delete)                | Automatic Garbage Collection        | Automatic Garbage Collection       |
| OOP           | Classes, multiple inheritance      | Classes, single inheritance (interfaces for more) | Classes, multiple inheritance |
| Portability   | Works only where compiled          | "Write once, run anywhere" (JVM)    | Runs anywhere with Python installed |
| Use Cases     | Games, OS, system programming      | Android apps, enterprise projects   | AI, data science, web apps, automation |
| Learning      | Hard (complex + memory handling)   | Medium (strict but consistent)      | Easy (simple syntax)               |

---

####### Why Speed Differs
- **C++** → Compiled directly to machine code → no middle layer → fastest  
- **Java** → Compiled to bytecode → then runs on JVM → slower than C++  
- **Python** → Interpreted line by line at runtime → slowest  

---

- **C++** = Fast (direct to machine)  
- **Java** = Medium (JVM layer)  
- **Python** = Slow (interpreter: An interpreter is a program that reads and executes code line by line directly, without converting it into machine code first.)

#####List
A list is an ordered and changeable (mutable) collection of items. Lists allow duplicate members.

Important Points:

- Ordered: The items have a defined order, and that order will not change.

- Mutable: You can change, add, and remove items in a list after it has been created.

- Allows Duplicates: A list can contain items with the same value.

- Syntax: Defined by square brackets [ ].

```python
# 1. Creating a list
fruits = ["apple", "banana", "cherry"]
print(fruits)
```

```python
# 2. Accessing items by index (starts at 0)
first_fruit = fruits[0]
print("First fruit:", first_fruit)
```

```python
# 3. Changing an item (Lists are mutable)
fruits[1] = "blackberry"
print("List after change:", fruits)
```

```python
# 4. Adding an item to the end
fruits.append("orange")
print("List after append:", fruits)
```

```python
# 5. Adding an item at a specific position
fruits.insert(1, "mango")
print("List after insert:", fruits)
```

```python
# 6. Removing an item
fruits.remove("cherry")
print("List after remove:", fruits)
```

```python
# 7. Getting the length of a list
print("Number of fruits:", len(fruits))
```

```python
# 8. A list can contain different data types, including duplicates
mixed_list = [1, "hello", 3.14, "hello", True]
print(mixed_list)
```

Important Points:

-  Ordered: Just like lists, items have a defined order.

- Immutable: Once a tuple is created, you cannot change, add, or remove items. This is the key difference from a list.

- Allows Duplicates: Tuples can have items with the same value.

- Syntax: Defined by round brackets ().

#### 1. Accessing Elements in Lists and Tuples
Lists and Tuples are both ordered, so they share the same methods for accessing elements based on their position (index).

```python
'''
a) Positive Indexing
Access a single item using its index, starting from 0 for the first item.
'''
my_list = ['p', 'y', 't', 'h', 'o', 'n']
my_tuple = ('p', 'y', 't', 'h', 'o', 'n')

# Access the first element (index 0)
print("First item of list:", my_list[0])  # Output: p
print("First item of tuple:", my_tuple[0]) # Output: p

```

```python
'''
b) Negative Indexing
Access items starting from the end of the collection. -1 refers to the last item, -2 to the second-to-last, and so on.
'''
# Access the last item
print("Last item of list:", my_list[-1]) # Output: n

# Access the second-to-last item
print("Second-to-last item of tuple:", my_tuple[-2]) # Output: o
```

```python
st1 = 'h'
st2 = "h"
print(type(st2))
print(st1 == st1)
```

```python
'''
c) Slicing
Access a range of items by specifying a start and end index ([start:stop]). The stop index is not included in the result.
'''
#p y t h o n
#
# Get items from index 2 up to (but not including) index 5
print("Slice [2:5]:", my_list[2:5]) # Output: ['t', 'h', 'o']
#   [Start : Stop]
# Get items from the beginning up to index 3
print("Slice [:3]:", my_list[:3]) # Output: ['p', 'y', 't']

# Get items from index 3 to the end
print("Slice [3:]:", my_list[3:]) # Output: ['h', 'o', 'n']

# Get every second item in the list (slicing with a step)
print("Slice [::2]:", my_list[::2]) # Output: ['p', 't', 'o']
# [Start : Stop : Step]
```

```python
name, age, cgpa = ["Alice", 25, 3.8]
print(name)
print(age)
print(cgpa)
```

```python
'''
d) Unpacking
Assign items from a list or tuple to multiple variables at once. The number of variables must match the number of items.
'''
person = ["Alice", 25, "Engineer"]
name, age, profession = person
print(name)      # Output: Alice
print(age)       # Output: 25
print(profession) # Output: Engineer

# Unpacking with a "catch-all" variable using *
numbers = [1, 2, 3, 4, 5]
first, second, *rest = numbers
print("First:", first) # Output: 1
print("Rest:", rest)  # Output: [3, 4, 5]
```

####2. Accessing Elements in Sets
Sets are unordered and unindexed, so you cannot access items by an index number.

```python
'''
a) Checking for Membership
The most common way to "access" an item in a set is to check if it exists using the in keyword. This operation is extremely fast and efficient for sets.

'''
my_set = {"apple", "banana", "cherry"}

# Check if an item exists
if "banana" in my_set:
    print("Yes, 'banana' is in the set.")
else:
    print("No, 'banana' is not in the set.")
```

```python
'''
b) Looping (Traversing)
To work with the elements, you must loop through the set.
'''
for item in my_set:
    print(item)
```

```python
'''
c) Converting to a List
If you need to access items by index (and the order doesn't matter), you can convert the set to a list first.
'''

set_as_list = list(my_set)
print("Set converted to list:", set_as_list)
print("First item of the converted list:", set_as_list[0])
```

####3. Accessing Elements in Dictionaries
Dictionaries use keys instead of indexes to access their corresponding values.

```python
'''
a) Using Square Brackets [key]
This is the most direct way to access a value. However, it will raise a KeyError if the key does not exist.
'''
student = {"name": "Bob", "age": 22, "major": "History"}

# Access the value associated with the 'name' key
print("Name:", student["name"]) # Output: Bob

# This will cause a KeyError because 'gpa' does not exist
# print(student["gpa"])
```

```python
'''
b) Using the .get() Method
This is a safer way to access a value. It returns the value for a key if it exists, but returns None (or a specified default value) if the key is not found. It does not raise an error.
'''
# Safely access the value for 'age'
age = student.get("age")
print("Age:", age) # Output: 22

# Safely access a key that does not exist
gpa = student.get("gpa")
print("GPA:", gpa) # Output: None

# Provide a default value if the key is not found
grade = student.get("grade", "N/A")
print("Grade:", grade) # Output: N/A
```

```python
'''
c) Accessing Keys, Values, and Items
You can retrieve collections of all keys, values, or key-value pairs.
'''
# Get all keys as a list-like view object
keys = student.keys()
print("All keys:", keys) # Output: dict_keys(['name', 'age', 'major'])

# Get all values
values = student.values()
print("All values:", values) # Output: dict_values(['Bob', 22, 'History'])

# Get all key-value pairs as tuples
items = student.items()
print("All items:", items) # Output: dict_items([('name', 'Bob'), ('age', 22), ('major', 'History')])
```

###SET
A set is an unordered, unchangeable, and unindexed collection of items. Sets do not allow duplicate members.

* Note: Set items are unchangeable, but you can add or remove items from the set itself.

Important Points:

- Unordered: The items in a set do not have a defined order.

- No Duplicates: Each item in a set must be unique.

- Mutable: You can add and remove items from a set.

- Syntax: Defined by curly braces {}.

```python
# 1. Creating a set (duplicates are automatically removed)
numbers = {1, 2, 3, 4, 3, 2}
print("Set of numbers:", numbers) # Output: {1, 2, 3, 4}
```

```python
# 2. Adding an item to a set
numbers.add(5)
print("Set after add:", numbers)
```

```python
# 3. Removing an item from a set
numbers.remove(3)
print("Set after remove:", numbers)
```

```python
# 4. Checking if an item is in a set (this is very fast!)
if 4 in numbers:
    print("4 is in the set.")
```

```python
# 5. Set Operations: Union (all items from both sets)
set_a = {1, 2, 3}
set_b = {3, 4, 5}
union_set = set_a | set_b
print("Union:", union_set)
```

```python
# 6. Set Operations: Intersection (items that are in both sets)
intersection_set = set_a & set_b
print("Intersection:", intersection_set)
```

```python
# create a set
my_set = {"apple", "banana", "cherry", 10, 20}

print("Traversing the set:")
for item in my_set:
    print(item)
```

```python
# Indexing not allowed
s = {1, 2, 3}
# print(s[0])   # Error: 'set' object is not subscriptable
```

```python
# Unhashable (mutable) items cannot be inside a set
s = {1, 2, [3, 4]}  #  Error: unhashable type: 'list'
```

```python
# Removing an item that does not exist (with remove)
s = {1, 2, 3}
s.remove(10)   # KeyError: 10 not in set
```

####Uses of Sets

```python
# Avoid duplicates
nums = [1, 2, 2, 3, 4, 4]
unique_nums = set(nums)
print(unique_nums)  # {1, 2, 3, 4}
```

```python
# Mathematical operations (union, intersection, difference)
a = {1, 2, 3}
b = {3, 4, 5}

print("Union:", a | b)         # {1, 2, 3, 4, 5}
print("Intersection:", a & b)  # {3}
print("Difference:", a - b)    # {1, 2}
```

```python
# Membership checking (fast lookup)
s = {10, 20, 30}
print(20 in s)   # True
print(100 in s)  # False
```

###Dictionaries
A dictionary is an ordered (in Python 3.7+) and changeable collection that stores data in key-value pairs.

Important Points:

- Key-Value Pairs: Each item has a key and a corresponding value.

- Unique Keys: Dictionary keys must be unique.

-  Mutable: You can change, add, and remove items.

- Syntax: Defined by curly braces {} with key: value pairs.

```python
# 1. Creating a dictionary
student = {
    "name": "Alice",
    "age": 21,
    "major": "Computer Science"
}
print(student)
```

```python
# 2. Accessing a value by its key
student_name = student["name"]
print("Student's name:", student_name)
```

```python
# 3. Changing a value
student["age"] = 22
print("Dictionary after change:", student)
```

```python
# 4. Adding a new key-value pair
student["gpa"] = 3.8
print("Dictionary after adding GPA:", student)
```

```python
# 5. Removing a key-value pair
del student["major"]
print("Dictionary after removing major:", student)
```

```python
# 6. Getting all keys
print("All keys:", student.keys())
```

```python
# 7. Getting all values
print("All values:", student.values())
```

```python
# 8. Checking if a key exists
if "name" in student:
    print("The key 'name' exists in the dictionary.")
```

| Feature | List `[]` | Tuple `()` | Set `{}` | Dictionary `{key:value}` |
| :--- | :--- | :--- | :--- | :--- |
| **Ordered** | Yes | Yes | No | Yes (Python 3.7+) |
| **Mutable** | Yes | No | Yes | Yes |
| **Duplicates Allowed** | Yes | Yes | No | No (for keys) |
| **How to Access** | Index (e.g., `my_list[0]`) | Index (e.g., `my_tuple[0]`) | No indexing | Key (e.g., `my_dict["key"]`) |
| **Use Case** | A collection of items you might need to change. | A collection of items that should never change. | Storing unique items and performing math set operations. | Storing related information as key-value pairs. |

```python
# Find the Smallest Number in a List
values = [4, 2, 9, 1, 7]
smallest = values[0]
for val in values:
    if val < smallest:
        smallest = val
print("The smallest value is:", smallest)
```
