#Matplotlib

Matplotlib is a data visualization library in Python.
- It allows us to create graphs, charts, and plots to represent data visually.
- Commonly used with NumPy or Pandas for data visualization.

````python
import matplotlib.pyplot as plt
````

#####Matplotlib Architecture

Matplotlib mainly has two interfaces:

- Pyplot Interface – A MATLAB-style state machine interface (plt.plot(), plt.show()).

- Object-Oriented Interface – Control with Figure and Axes objects.

````python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()


# plt.plot() : draws the line graph.

# plt.title(), plt.xlabel(), plt.ylabel() : add labels.

# plt.show() : displays the plot.
````

#####Plot Customization

````python
import matplotlib.pyplot as plt
x = ["10th", "12th", "1st sem", "2nd Sem"]
y = [85, 73, 92, 86]
# - , -- , -. , :
plt.plot(x, y, color='blue', linestyle='-.', marker='o', markerfacecolor='red', linewidth=1)
plt.title("Customized Line Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.grid(True)
plt.show()
````

#####Bar Chart

````python
students = ['A', 'B', 'C', 'D', 'E']
marks = [85, 70, 90, 60, 75]

plt.bar(students, marks, color='orange')
plt.title("Student Marks Bar Chart")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
````

#####Pie Chart

````python
languages = ['Python', 'C++', 'Java', 'C', 'JS']
popularity = [35, 20, 25, 10, 10]

plt.pie(popularity, labels=languages, autopct='%1.2f%%', startangle=0)
plt.title("Programming Language Popularity")
plt.show()
````

````python
# Data
browsers = ['Chrome', 'Firefox', 'Edge', 'Brave', 'Others']
market_share = [64, 19, 9, 5, 3]

# Explode the first slice to highlight it
explode = [0.1, 0, 0, 0, 0]

# Create pie chart
plt.pie(market_share,
        labels=browsers,
        explode=explode,
        autopct='%1.1f%%',
        startangle=90,
        shadow=True)

plt.title("Web Browser Market Share 2025")
plt.show()
````

#####Scatter Plot

````python
x = [5, 7, 8, 10, 12, 15, 18]
y = [12, 16, 20, 22, 25, 30, 35]

plt.scatter(x, y, color='purple', marker='*')
plt.title("Scatter Plot Example")
plt.xlabel("Independent Variable")
plt.ylabel("Dependent Variable")
plt.show()
````

#####Multiple Plots

````python
x = [1, 2, 3, 4, 5]

plt.plot(x, [a**2 for a in x], label='x^2')
plt.plot(x, [a**3 for a in x], label='x^3')
plt.title("Multiple Plots Example")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
````

#####Subplots

````python
x = [1, 2, 3, 4, 5]

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(x, [a**2 for a in x], 'r--')
plt.title("Square Function")

plt.subplot(1, 2, 2)
plt.plot(x, [a**3 for a in x], 'b-.')
plt.title("Cube Function")

plt.tight_layout()
plt.show()
````

|  **Matplotlib Function** |  **Description** |
|:----------------------------|:------------------|
| `plot()`                    | Draws a simple line graph |
| `bar()`                     | Creates a vertical bar chart |
| `pie()`                     | Creates a circular pie chart |
| `scatter()`                 | Plots data points on an X-Y plane |
| `xlabel()` / `ylabel()`     | Adds labels to X and Y axes |
| `title()`                   | Adds a title to the plot |
| `legend()`                  | Displays a legend for plotted data |
| `subplot()`                 | Creates multiple plots in one figure |


#Numpy


### What is NumPy?

NumPy stands for **Numerical Python**.  
It provides fast mathematical operations on large data sets using **arrays** instead of slow Python lists.

- Core object: **ndarray** (n-dimensional array).
- Used heavily in:
  - Data Science
  - Machine Learning
  - Image Processing
  - Mathematical Modeling

````python
import numpy as np
````

| Feature         | Python List   | NumPy Array                    |
| :-------------- | :------------ | :----------------------------- |
| Data Type       | Can mix types | Same type only                 |
| Speed           | Slower        | Much faster                    |
| Memory          | Uses more     | Efficient                      |
| Math Operations | Manual (loop) | Vectorized (direct operations) |


````python
# Performance Difference
import numpy as np
import time

size = 1_000_000
L = range(size)
A = np.arange(size)

start = time.time()
[ i*2 for i in L ]
print("Python list:", time.time()-start)

start = time.time()
A*2
print("NumPy array:", time.time()-start)
````

###Creating NumPy Arrays

````python
# From Python list
arr = np.array([1, 2, 3, 4, 5])
print(arr)
````

| Function                        | Description                   | Example                            |
| :------------------------------ | :---------------------------- | :--------------------------------- |
| `np.array()`                    | Creates array from list/tuple | `np.array([1,2,3])`                |
| `np.zeros(n)`                   | Array of zeros                | `np.zeros(5)` → `[0. 0. 0. 0. 0.]` |
| `np.ones(n)`                    | Array of ones                 | `np.ones(5)`                       |
| `np.arange(start, stop, step)`  | Range of numbers              | `np.arange(0,10,2)`                |
| `np.linspace(start, stop, num)` | Evenly spaced values          | `np.linspace(0,1,5)`               |
| `np.eye(n)`                     | Identity matrix               | `np.eye(3)`                        |
| `np.random.rand(n)`             | Random numbers (0–1)          | `np.random.rand(5)`                |


````python
a = np.zeros(4)
b = np.ones(3)
c = np.arange(2, 10, 2)
d = np.linspace(1, 5, 5)

print(a)
print(b)
print(c)
print(d)
````

###Array Attributes

````python
arr = np.array([[1,2,3],[4,5,6]])

print("Shape:", arr.shape)
print("Size:", arr.size)
print("Dimensions:", arr.ndim)
print("Data type:", arr.dtype)
````

| Attribute | Meaning                             |
| :-------- | :---------------------------------- |
| `.shape`  | Rows × Columns                      |
| `.size`   | Total elements                      |
| `.ndim`   | Number of dimensions                |
| `.dtype`  | Type of elements (int, float, etc.) |


###Array Indexing and Slicing

````python
# 1-D
arr = np.array([10,20,30,40,50])

print(arr[0])   # first element
print(arr[-1])  # last element
print(arr[1:4]) # slice
````

````python
# 2-D
matrix = np.array([[10,20,30],
                   [40,50,60],
                   [70,80,90]])

print(matrix[0,1])   # element at row 0, col 1
print(matrix[:,1])   # entire 2nd column
print(matrix[1,:])   # entire 2nd row
````

###Basic Array Operations

````python
a = np.array([1,2,3])
b = np.array([4,5,6])

print("Add:", a + b)
print("Subtract:", a - b)
print("Multiply:", a * b)
print("Divide:", a / b)
print("Power:", a ** 2)
````

| Function       | Description        | Example          |
| :------------- | :----------------- | :--------------- |
| `np.min(a)`    | Minimum value      | `np.min(arr)`    |
| `np.max(a)`    | Maximum value      | `np.max(arr)`    |
| `np.sum(a)`    | Sum of elements    | `np.sum(arr)`    |
| `np.mean(a)`   | Mean (average)     | `np.mean(arr)`   |
| `np.median(a)` | Median value       | `np.median(arr)` |
| `np.std(a)`    | Standard deviation | `np.std(arr)`    |


````python
arr = np.array([10,20,30,40,50])
print("Sum:", np.sum(arr))
print("Mean:", np.mean(arr))
print("Std Dev:", np.std(arr))
````

###Reshaping Arrays

````python
arr = np.arange(1, 9)
reshaped = arr.reshape(2,4)
print(reshaped)
````

| Method           | Description                |
| :--------------- | :------------------------- |
| `.reshape(r, c)` | Changes array shape        |
| `.flatten()`     | Converts 2D → 1D           |
| `.T`             | Transpose (rows ↔ columns) |


````python
arr = np.arange(5,11)
rr = arr.reshape(2,3)
print(rr)
````

###Joining and Splitting

````python
a = np.array([1,2,3])
b = np.array([4,5,6])

print(np.concatenate((a,b)))
print(np.split(np.arange(10), 2))
````

###Multidimensional Example

````python
arr = np.arange(1, 10).reshape(3,3)
print("Array:\n", arr)
print("Transpose:\n", arr.T)
print("Sum across rows:", arr.sum(axis=1))
print("Sum across cols:", arr.sum(axis=0))
````

#Panda

Pandas is a Python library used for **data manipulation** and **analysis**.  
It provides two main data structures:

- **Series**: 1D labeled array  
- **DataFrame**: 2D table (rows + columns), like an Excel sheet

Pandas is built on top of NumPy and is widely used in:
- Data science
- Machine learning
- Data preprocessing
- Data cleaning
- Exploratory data analysis

Importing Pandas:

```python
import pandas as pd

##Pandas Data Structures

###Series

A Series is a one-dimensional labeled array.

````python
import pandas as pd

s = pd.Series([10, 20, 30, 40])
print(s)
````

Creating series with custom index:

````python
s = pd.Series([10,20,30], index=['a','b','c'])
````

###DataFrame
A DataFrame is a 2D table of rows and columns.

````python
data = {
    'Name': ['Aman', 'Raj', 'Raju'],
    'Marks': [85, 90, 95],
    'Age': [20, 21, 22]
}

df = pd.DataFrame(data)
print(df)
````

##Reading and Writing Data
Pandas can read many file formats.

####Reading files

````python
df = pd.read_csv('data.csv')
# df = pd.read_excel('data.xlsx')
# df = pd.read_json('data.json')
````

####Writing files

````python
df.to_csv('output.csv', index=False)
# df.to_excel('output.xlsx', index=False)
# df.to_json('output.json')
````

##Basic DataFrame Operations

###View Data

````python
df.head()      # first 5 rows
df.tail()      # last 5 rows
df.info()      # column info, datatype
df.describe()  # statistics summary
````

###Selecting Rows and Columns

####Selecting columns

````python
df['Name']
df[['Name', 'Age']]
````

####Selecting rows

````python
df.iloc[0]      # row by index
df.loc[1]       # row by label
df.loc[0:2]     # multiple rows
````

````python
#Selecting specific cell
df.loc[1, 'Marks']
````

##Filtering Data

````python
df[df['Marks'] > 85]
df[(df['Marks'] > 85) & (df['Age'] < 22)]
````

##Adding, Updating, and Deleting Columns

````python
# Add Col
df['Result'] = ['Pass', 'Pass', 'Pass']

# Update Col
df['Marks'] = df['Marks'] + 5

# Del Col
df.drop('Age', axis=1, inplace=True)
````

##Handling Missing Data

````python
# Detect missing values
df.isnull()
df.isnull().sum()
````

````python
# Replace missing values
df.fillna(0)
df.fillna(method='ffill')  # forward fill
df.fillna(method='bfill')  # backward fill
````

````python
# Remove missing rows
df.dropna()
````

##Sorting and Describing Data

````python
# Sort
df.sort_values('Marks')
# df.sort_values('Marks', ascending=False)
````

###Grouping Data (groupby)

````python
df.groupby('Class')['Marks'].mean()


# Useful for:
# finding average marks per class
# total sales per month
# count per category
````

###Merging, Joining, and Concatenation

````python
# Concatenate
# pd.concat([df1, df2])

# Merge
# pd.merge(df1, df2, on='ID')

# Apply Functions to Columns
df['Marks'] = df['Marks'].apply(lambda x: x + 5)
````
