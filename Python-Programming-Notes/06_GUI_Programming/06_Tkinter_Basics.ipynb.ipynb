````python
from tkinter import *

def click_button(value):
    entry.insert(END, value)

def clear():
    entry.delete(0, END)

def calculate():
    try:
        result = eval(entry.get())
        entry.delete(0, END)
        entry.insert(END, str(result))
    except:
        entry.delete(0, END)
        entry.insert(END, "Error")

root = Tk()
root.title("Simple Calculator")
root.geometry("950x980")
root.configure(bg="#e8e8e8")
root.resizable(False, False)

# Display
entry = Entry(root, font=("Arial", 28), borderwidth=8, relief=RIDGE, justify='right', bg="white")
entry.grid(row=0, column=0, columnspan=4, pady=20, padx=10, ipady=15)

button_style = {
    "font": ("Arial", 22),
    "width": 4,
    "height": 2,
    "borderwidth": 3,
    "relief": RIDGE,
    "bg": "#f2f2f2"
}

buttons = [
    ('7', 1, 0), ('8', 1, 1), ('9', 1, 2), ('/', 1, 3),
    ('4', 2, 0), ('5', 2, 1), ('6', 2, 2), ('*', 2, 3),
    ('1', 3, 0), ('2', 3, 1), ('3', 3, 2), ('-', 3, 3),
    ('0', 4, 0), ('.', 4, 1), ('=', 4, 2), ('+', 4, 3),
]

for (text, r, c) in buttons:
    if text == "=":
        Button(root, text=text, command=calculate, bg="#4CAF50", fg="white",
               font=("Arial", 22), width=4, height=2, borderwidth=3, relief=RIDGE
        ).grid(row=r, column=c, padx=8, pady=10)
    else:
        Button(root, text=text, command=lambda t=text: click_button(t), **button_style
        ).grid(row=r, column=c, padx=8, pady=10)

# Clear Button (big and red)
Button(root, text="CLEAR", command=clear,
       font=("Arial", 22), width=18, height=2,
       bg="#d9534f", fg="white", borderwidth=3, relief=RIDGE
).grid(row=5, column=0, columnspan=4, pady=15)

root.mainloop()
````

## Layout Managers
Used to position widgets.

````python
# pack()
# Simple layout manager.
Label(root, text="Top").pack(side=TOP)
Label(root, text="Bottom").pack(side=BOTTOM)
Label(root, text="Left").pack(side=LEFT)
Label(root, text="Right").pack(side=RIGHT)
````

````python
# grid()
# Best for form-like layouts.
Label(root, text="Name").grid(row=0, column=0)
Entry(root).grid(row=0, column=1)
````

````python
# place()
# Absolute positioning.

Label(root, text="Hello").place(x=100, y=50)
````

## Event Handling
Run functions when the user interacts.

````python
# Button event
Button(root, text="Click", command=myfunc).pack()
````

````python
# Keyboard event
def keypress(e):
    print("Pressed:", e.char)

root.bind("<Key>", keypress)
````

````python
# Mouse event
def click(e):
    print("Mouse clicked at", e.x, e.y)

root = Tk()
root.title("My GUI")
root.geometry("400x300")
root.bind("<Button-1>", click)
root.mainloop()
````

## Frames
Frames are containers used to group widgets.

````python
frame = Frame(root, bg="lightblue")
frame.pack()

Button(frame, text="OK").pack(side=LEFT)
Button(frame, text="Cancel").pack(side=LEFT)
````

## Menus in Tkinter

````python
menu = Menu(root)
root.config(menu=menu)

file_menu = Menu(menu, tearoff=0)
menu.add_cascade(label="File", menu=file_menu)

file_menu.add_command(label="New")
file_menu.add_command(label="Open")
file_menu.add_separator()
file_menu.add_command(label="Exit", command=root.quit)
````
